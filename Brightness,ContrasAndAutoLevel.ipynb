```python
#import library
import cv2
import imageio
import numpy as np
import matplotlib.pyplot as plt 
```

```python
#Masukkan gambar dan menamppilkan nya
img = imageio.imread("orange.jpg")
plt.imshow(img)
plt.show()
```

```python
#define reso dan tipe gambar
img_height = img.shape[0] # Mengambil tinggi dari gambar
img_width = img.shape[1] # Mengambil lebar dari gambar
img_channel = img.shape[2] # Mengambil warna dari gambar
img_type = img.dtype # Mengambil type gambar
```

```python
#membuat brightness untuk gambar grayscale
#membuat variable img_brightness untuk menampung
img_brightness = np.zeros(img.shape, dtype=np.uint8)
```

```python
def brighter(nilai): # fungsi brighter dengan parameter nilai
    for y in range(0, img_height): # Looping untuk tinggi gambar 
        for x in range(0, img_width): # Looping untuk lebar gambar
            red = img[y][x][0]
            green = img[y][x][1] # untuk mendapatkan informasi piksel gambar dari RGB
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 # menghitung nilai rata-rata untuk mendapatkan nilai keabuan
            gray += nilai
            if gray > 255: # nilai batas
                gray = 255
            if gray < 0: # nilai batas
                gray = 0
            img_brightness[y][x] = (gray, gray, gray) # untuk menampung nilai baru
```

```python
#menampilkan gambar dengan parameter -100 dan +100
fig, axes = plt.subplots(1, 2, figsize=(8, 8)) 
ax = axes.ravel() 

brighter(-100)
ax[0].imshow(img_brightness) 
ax[0].set_title("Brightness -100") 

brighter(100)
ax[1].imshow(img_brightness) 
ax[1].set_title("Brightness +100") 

fig.tight_layout
plt.show()
```

```python
#contrast
img_contrass = np.zeros(img.shape, dtype=np.uint8)
```

```python
def contrass(nilai): # fungsi contrass dengan parameter nilai
    for y in range(0, img_height): # Looping tinggi gambar
        for x in range(0, img_width): # Looping lebar gambar
            red = img[y][x][0]
            green = img[y][x][1] # untuk mendapatkan informasi piksel gambar dari RGB
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3 # menghitung nilai rata-rata untuk mendapatkan nilai keabuan
            gray *= nilai # mengalikan nilai keabuan dengan nilai
            if gray > 255: # nilai batas
                gray = 255
            img_contrass[y][x] = (gray, gray, gray) #  untuk menampung nilai baru
```

```python
#menampilkan gambar dengan parameter contrass -2 dan +200
fig, axes = plt.subplots(1, 2, figsize=(8, 8)) 
ax = axes.ravel() 

contrass(2)
ax[0].imshow(img_contrass) 
ax[0].set_title("Contrass 2") 

contrass(15)
ax[1].imshow(img_contrass) 
ax[1].set_title("Contrass 15") 

fig.tight_layout
plt.show()
```

```python
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)
```

```python
def autocontrass():
    xmax = 255
    xmin = 0
    d = 0
    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada hitungan
    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            if gray < xmax:
                xmax = gray
            if gray > xmin:
                xmin = gray
    d = xmin-xmax
    for y in range(0, img_height):
        for x in range(0, img_width):
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            gray = (int(red) + int(green) + int(blue)) / 3
            gray = int(float(255/d) * (gray-xmax))
            img_autocontrass[y][x] = (gray, gray, gray)
```

```python
autocontrass()
plt.imshow(img_autocontrass)
plt.title("Contrass Autolevel")
plt.show()
```
